```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Let's bring in the DataFrame
df = pd.read_csv("titanic_features.csv")
df.head()
```

```python
#Now, let's start with the first task, using Histograms or KDE plots for 3 numerical features
#I will doing the code similair to Day 26 - Data Distributions
#First, Age.
#Hist
plt.figure(figsize=(6, 4))
df["Age"].hist(bins=30, edgecolor="black")
plt.title("Age Distribution — Histogram")
plt.xlabel("Age"); plt.ylabel("Amount")
plt.tight_layout(); plt.show()
#KDE
plt.figure(figsize=(6, 4))
sns.kdeplot(df["Age"], fill=True)
plt.title("Age Distribution — KDE")
plt.xlabel("Age"); plt.ylabel("Density")
plt.tight_layout(); plt.show()
```

```python
"""Both the Histogram and the KDE Plot show us that the majority 
of the passengers' ages is around 30, with the highest age being around 80 or more.
Kids also made a big amount of the density; from my analysis alone, they could be the 
third biggest age group after Adults and Young Adults and before Seniors.
It seems slightly right-skewed slightly bimodal"""
```

```python
#For the second numerical features I'll be choosing..
#Family Size
#Hist
plt.figure(figsize=(6, 4))
df["family_size"].hist(bins=10, edgecolor="black")
plt.title("Family Size Distribution — Histogram")
plt.xlabel("family_size"); plt.ylabel("Amount")
plt.tight_layout(); plt.show()
#KDE
plt.figure(figsize=(6, 4))
sns.kdeplot(df["family_size"], fill=True)
plt.title("Family Size Distribution — KDE")
plt.xlabel("family_size"); plt.ylabel("Density")
plt.tight_layout(); plt.show()
```

```python
"""We can see from both charts that most people went alone aboard the Titanic, which is
very sad considering the fact some never got to see their families again. Moreover, we
can see that the largest size brought aboard the Titanic is ~11, and we can only hope
that the entire family made it out alive.
It looks heavily right-skewed."""
```

```python
#And finally, the third numerical features will be...
#Fare for Person (Why I chose this one out of the other two 
#specifically is because I saw that showing how each person paid for 
#themselves is better than showing what they payed overall)
#Hist
plt.figure(figsize=(6, 4))
df["fare_per_person"].hist(bins=30, edgecolor="black")
plt.title("Fare/Person Distribution — Histogram")
plt.xlabel("fare_per_person"); plt.ylabel("Amount")
plt.tight_layout(); plt.show()
#KDE
plt.figure(figsize=(6, 4))
sns.kdeplot(df["fare_per_person"], fill=True)
plt.title("Fare/Person Distribution — KDE")
plt.xlabel("fare_per_person"); plt.ylabel("Density")
plt.tight_layout(); plt.show()
```

```python
"""The charts look Bimodal to Multimodal, and right-skewed at the same time.
The majority paid 0-20 for their fare, with the most reaching almost 70.
20 to 40 also has a bulge, which makes it coincide with the passenger classes as it 
clusters into 3 groups."""
```

```python
#Now, next task is: Grouped boxplots: compare the target across at least 2 categories
#Since I didn't what the task needed, nor how to satisfy the task, I had to ask Gemini to help me do it.
#Both Code and Paragraph by Gemini:

# 1. Setup the figure structure
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 2. Create Grouped Boxplot 1: Fare vs Pclass, grouped by Survived
sns.boxplot(ax=axes[0], data=df, x='Pclass_ord', y='Fare_transformed', 
            hue='Survived', palette='Set1')
axes[0].set_title('Fare Distribution: Pclass vs Survived')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Fare (Transformed)')

# 3. Create Grouped Boxplot 2: Age vs Pclass, grouped by Survived
sns.boxplot(ax=axes[1], data=df, x='Pclass_ord', y='Age', 
            hue='Survived', palette='Set1')
axes[1].set_title('Age Distribution: Pclass vs Survived')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Age')

plt.tight_layout()
plt.show()
```

```python
"""The charts exhibit multimodal distributions, with Fare being heavily right-skewed and Age showing distinct density peaks across groups.
Fare clusters tightly across passenger classes, where the highest class clearly maintains a higher survival threshold compared to the lower 
tiers. Age similarly clusters, revealing a younger demographic in lower classes that, when segmented by survival, shows distinct variance 
in outcomes compared to older, higher-class counterparts."""
```

```python
#Now, Correlation heatmap: show the top 10 features most correlated with the target

#Let's Select numeric features as per Day 35
numeric_df = df.select_dtypes(include=[np.number])

corr_matrix = numeric_df.corr()

#Now we identify Top 10 features correlated with 'Survived' (Target)
#We use abs() to find the strongest relationships, whether positive or negative
top_10_list = corr_matrix['Survived'].abs().sort_values(ascending=False)[1:11].index.tolist()
plot_cols = ['Survived'] + top_10_list

#Now we can make the heatmap!
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix.loc[plot_cols, plot_cols], 
            annot=True, 
            cmap='coolwarm', 
            fmt=".2f", 
            vmin=-1, 
            vmax=1, 
            linewidths=0.5)

plt.title('Top 10 Features Correlated with Survival')
plt.show()
```

```python
"""Ignoring the obvious full correlation with the target (Survived), we can see other high correlations 
in this chart. For example, 'fare_per_person' has a 0.70 correlation to 'Pclass_ord', which aligns to how 
paying more gets you into a higher class. Other examples of high correlation in 'family_size' with 'Parch'
and 'SibSp'"""
```

```python
#Next up, we make the scatterplot.

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='Set1', alpha=0.6)

plt.title('Relationship: Age vs. Fare by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
"""The chart above shows a cluster in Fare 0-20 and Age 10-50, and the color meaning that people at
the lowest class mostly died, and we can see that the highest points in the  plot, the rich ones, mostly
survived. And finally, the middle class shows a mix of both dead people and survivors, so here the sex 
might've mattered more than Age or Fare."""
```

```python
#Last task, compute mean target by a category.

#I'll compute the target's mean by passenger class.
survival_by_class = df.groupby('Pclass_ord')['Survived'].mean().sort_values()

print("Mean Survival Rate by Class:")
print(survival_by_class)

# We identify highest and lowest survival rates here.
print(f"\nHighest Survival: Class {survival_by_class.idxmax()}")
print(f"Lowest Survival: Class {survival_by_class.idxmin()}")
```
