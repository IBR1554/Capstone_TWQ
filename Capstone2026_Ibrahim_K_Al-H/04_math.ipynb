```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic_features.csv")
df.head()
```

```python
#First task, we compute the std and mean of the target (Survived)

mean = np.sum(df["Survived"])/ len(df["Survived"])
std = np.sqrt(np.sum((df["Survived"] - mean)**2) / len(df["Survived"]))
#formula for std gotten from ChatGPT because I couldn't find it on the course slides

print(f"Mean: {mean} \nStandard Deviation: {std}")
```

```python
#Next, broadcasting
#I choose Age, because it is numeric
from sklearn.preprocessing import StandardScaler 
X = df[["Age"]]
X_mean = np.sum(df["Age"])/ len(df["Age"])
X_std = np.sqrt(np.sum((df["Age"] - X_mean)**2) / len(df["Age"]))
Z = (X - X_mean) / X_std
#We bring in Standard Scaler so that we can compare
scaler = StandardScaler()
z_scaled = scaler.fit_transform(X)
print(f"Mean : {X_mean}\nSTD : {X_std}\nFirst 5 Manual standardizations:\n{Z[:5]}\nFirst 5 Standardizations using sklearn\n{z_scaled[:5]}")
#I had to use Ai for defining X because I didn't know how to make it work with SS but turns out double
#brackets was the solution
```

```python
#Final Task! time to estimate probability from the dataset
#I'll use the example: What's your possibility of surviving, given your sex?

# Probability of survival given Sex = female
p_female = df[df["Sex_female"] == True]["Survived"].mean()

# Probability of survival given Sex = male
p_male = df[df["Sex_female"] == False]["Survived"].mean()

print("P(Survived | Female):", p_female)
print("P(Survived | Male):", p_male)
#Idea from GPT
```
